# Question 5: Linear Regression with SKlearn
- Build a simple linear regression model to predict `tip` from `total_bill` using scikit-learn. 
- Calculate and interpret the `R-squared` value, and `MSE` for both the <b><font color='red'> training and testing sets</b></font>. 
- Generate a `residual` plot (using the testing dataset: y_test - y_pred) and analyze its distribution. 
- Interpret the `slope` and `intercept` of the regression line.
- Split the dataset into `70% for model training` and `30% for testing the model` 


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Split the dataset into 70% for model training and 30% for testing the model
X = tips.drop(columns=['sex','smoker','day','time','size', 'tip'])
y = tips.tip
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, train_size=0.7, 
                                                    random_state=123)
X_train.shape[0]+X_test.shape[0] == tips.shape[0]
X
```

```python
#Build a simple linear regression model to predict tip from total_bill using scikit-learn.
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Interpret the slope and intercept of the regression line
print('Intercept is:\t {}'.format(lm.intercept_))
print('Slopes are:\t {}'.format(lm.coef_))
```

```python
# Generate a residual plot (using the testing dataset: y_test - y_pred) and analyze its distribution.

y_pred = lm.predict(X_test) # These are the predicted values of the testing dataset
compare = pd.DataFrame({'y_test':y_test, 'y_pred':y_pred})
compare = compare.round(2)
# display(compare.head())
compare['delta'] = compare.y_test - compare.y_pred
compare.reset_index(inplace=True, drop=True)
display(compare.head())

fig, ax = plt.subplots(1,1, figsize = (15,7))
ax = plt.bar(x = compare.index, height=compare.delta)
sns.despine()
plt.title('The Residual with the testing dataset between the y_test - y_pred')
plt.xlabel('Index of the y_test value')
plt.ylabel('Residual')
```

```python
# Calculate and interpret the R-squared value, and MSE for both the training and testing sets.
print('R-squared on the training sample:\t {}'.format(lm.score(X_train, y_train)))
print('R-squared on the testing sample:\t {}'.format(lm.score(X_test, y_test)))

y_train_pred = lm.predict(X_train)
lin_train_mse = mean_squared_error(y_train, y_train_pred)
print('MSE on the training sample:\t {}'.format(lin_train_mse))

y_test_pred = lm.predict(X_test)
lin_test_mse = mean_squared_error(y_test, y_test_pred)
print('MSE on the training sample:\t {}'.format(lin_test_mse))

lin_train_r2 = r2_score(y_train, y_train_pred)
print('R-squared on the training sample from sklearn r2_score:\t {}'.format(lin_train_r2))

lin_test_r2 = r2_score(y_test, y_test_pred)
print('R-squared on the training sample from sklearn r2_score:\t {}'.format(lin_test_r2))
```

# Question 6: Multiple Linear Regression Model
- Build a multiple linear regression model to predict `tip` using `total_bill`, `size`, `sex`, `smoker`, and `day`. 
- Use the same split as above (`70% for training` and `30% for testing`).
- Evaluate the model's performance using the same metrics as in the simple linear regression (`R^2` & `MSE`).
- Compare the performance of the `simple` and `multiple` linear regression models.

```python
print('unique values for sex: ', tips.sex.unique())
print('unique values for smoker: ', tips.smoker.unique())
print('unique values for day: ', tips.day.unique())
print('unique values for time: ', tips.time.unique())
```

```python
#create dataframe to have columns needed for multiple linear regression model
X2 = tips
sex_results = X2.sex.tolist()
female = [1 if result == 'Female' else 0 for result in sex_results]
X2['sex_Female'] = female

smoker_results = X2.smoker.tolist()
smoker = [0 if result == 'No' else 1 for result in smoker_results]
X2['smoker_No'] = smoker

day_results = X2.day.tolist()
day_fri = [1 if result == 'Fri' else 0 for result in day_results]
day_sat = [1 if result == 'Sat' else 0 for result in day_results]
day_sun = [1 if result == 'Sun' else 0 for result in day_results]
X2['day_Fri'] = day_fri
X2['day_Sat'] = day_sat
X2['day_Sun'] = day_sun
#no need for thursday because if all day columns are 0, that means its thursday

time_results = X2.time.tolist()
time = [1 if result == 'Dinner' else 0 for result in time_results]
X2['time_Dinner'] = time

X2 = X2.drop(columns=['sex','smoker','day','time','size', 'tip'])
X2
```

```python
#Split the dataset into 70% for model training and 30% for testing the model
X = X2
y = tips.tip
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, train_size=0.7, 
                                                    random_state=123)
X_train.shape[0]+X_test.shape[0] == tips.shape[0]
X
```

```python
#Build a multiple linear regression model to predict tip using total_bill, size, sex, smoker, and day.
lm2 = LinearRegression()
lm2.fit(X_train,y_train)
```

```python
# Evaluate the model's performance using the same metrics as in the simple linear regression (R^2 & MSE).

# Interpret the slope and intercept of the regression line
print('Intercept is:\t {}'.format(lm2.intercept_))
print('Slopes are:\t {}'.format(lm2.coef_))

# Generate a residual plot (using the testing dataset: y_test - y_pred) and analyze its distribution.

y_pred = lm2.predict(X_test) # These are the predicted values of the testing dataset
compare = pd.DataFrame({'y_test':y_test, 'y_pred':y_pred})
compare = compare.round(2)
# display(compare.head())
compare['delta'] = compare.y_test - compare.y_pred
compare.reset_index(inplace=True, drop=True)
display(compare.head())

fig, ax = plt.subplots(1,1, figsize = (15,7))
ax = plt.bar(x = compare.index, height=compare.delta)
sns.despine()
plt.title('The Residual with the testing dataset between the y_test - y_pred')
plt.xlabel('Index of the y_test value')
plt.ylabel('Residual')

# Calculate and interpret the R-squared value, and MSE for both the training and testing sets.
print('R-squared on the training sample:\t {}'.format(lm2.score(X_train, y_train)))
print('R-squared on the testing sample:\t {}'.format(lm2.score(X_test, y_test)))

y_train_pred = lm2.predict(X_train)
lin_train_mse = mean_squared_error(y_train, y_train_pred)
print('MSE on the training sample:\t {}'.format(lin_train_mse))

y_test_pred = lm2.predict(X_test)
lin_test_mse = mean_squared_error(y_test, y_test_pred)
print('MSE on the training sample:\t {}'.format(lin_test_mse))

lin_train_r2 = r2_score(y_train, y_train_pred)
print('R-squared on the training sample from sklearn r2_score:\t {}'.format(lin_train_r2))

lin_test_r2 = r2_score(y_test, y_test_pred)
print('R-squared on the training sample from sklearn r2_score:\t {}'.format(lin_test_r2))
```

# Question 7: Regularization
- Apply `StandardScaler` on the datasets.
- Check which alpha parameter is the `optimal value` among [0.1, 1, 10, 100, 200]
- Apply `Ridge` Linear Regression with the `best alpha` value.
- What is your conclusion?

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge

# Apply StandardScaler on the datasets.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.astype(np.float64))
X_test_scaled = scaler.fit_transform(X_test.astype(np.float64))
# y_train_scaled = scaler.fit_transform(y_train.astype(np.float64))
# y_test_scaled = scaler.fit_transform(y_test.astype(np.float64))
X_train_scaled
X_test_scaled
# y_train_scaled
# y_test_scaled
```

```python
# Check which alpha parameter is the optimal value among [0.1, 1, 10, 100, 200]
alpha = [0.1, 1, 10, 100, 200]
for a in alpha:
    print('alpha=', a)
    ridge_reg = Ridge(alpha=a,solver="cholesky")
    ridge_reg.fit(X_train_scaled,y_train)
    y_pred_ridge_reg = ridge_reg.predict(X_test)
    compare = pd.DataFrame({'y_test':y_test, 'y_pred_ridge_reg':y_pred_ridge_reg})
    compare = compare.round(2)
    # display(compare.head())
    compare['delta'] = compare.y_test - compare.y_pred_ridge_reg
    compare.reset_index(inplace=True, drop=True)
    display(compare.head())


#alpha 200 is the best
```

```python
optimal_ridge_reg = Ridge(alpha=200,solver="cholesky")
optimal_ridge_reg.fit(X_train_scaled,y_train)
y_pred_ridge_reg = optimal_ridge_reg.predict(X_test)
compare = pd.DataFrame({'y_test':y_test, 'y_pred_ridge_reg':y_pred_ridge_reg})
compare = compare.round(2)
# display(compare.head())
compare['delta'] = compare.y_test - compare.y_pred_ridge_reg
compare.reset_index(inplace=True, drop=True)
display(compare.head())


fig, ax = plt.subplots(1,1, figsize = (15,7))
ax = plt.bar(x = compare.index, height=compare.delta)
sns.despine()
plt.title('The Residual with the testing dataset between the y_test - y_pred')
plt.xlabel('Index of the y_test value')
plt.ylabel('Residual')

# Calculate and interpret the R-squared value, and MSE for both the training and testing sets.
print('R-squared on the training sample:\t {}'.format(optimal_ridge_reg.score(X_train, y_train)))
print('R-squared on the testing sample:\t {}'.format(optimal_ridge_reg.score(X_test, y_test)))

y_train_pred = optimal_ridge_reg.predict(X_train_scaled)
lin_train_mse = mean_squared_error(y_train, y_train_pred)
print('MSE on the training sample:\t {}'.format(lin_train_mse))

y_test_pred = optimal_ridge_reg.predict(X_test_scaled)
lin_test_mse = mean_squared_error(y_test, y_test_pred)
print('MSE on the training sample:\t {}'.format(lin_test_mse))

lin_train_r2 = r2_score(y_train, y_train_pred)
print('R-squared on the training sample from sklearn r2_score:\t {}'.format(lin_train_r2))

lin_test_r2 = r2_score(y_test, y_test_pred)
print('R-squared on the training sample from sklearn r2_score:\t {}'.format(lin_test_r2))
```
